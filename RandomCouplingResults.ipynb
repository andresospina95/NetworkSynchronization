# Results

Takes the files generated by *RandomCoupling.py* and plots its synchronization diagram

```python
import numpy as np
import scipy.stats as ss
import matplotlib.pylab as plt
```

```python
s2K = np.load('s2K.npy')
s2r = np.load('s2r.npy')
s2rstd = np.load('s2rstd.npy')
```

```python
s3K = np.load('s3K.npy')
s3r = np.load('s3r.npy')
s3rstd = np.load('s3rstd.npy')
```

```python
s4K = np.load('s4K.npy')
s4r = np.load('s4r.npy')
s4rstd = np.load('s4rstd.npy')
```

```python
unwK = np.load('unwK.npy')
unwr = np.load('unwr.npy')
unwrstd = np.load('unwrstd.npy')
```

```python
fig, ax = plt.subplots(1,1, figsize=(9, 6.))

# ax.set_title('SL network with Gamma distributed coupling strength')
ax.set_xlabel(r'$\langle \lambda \rangle$', fontsize=20)
ax.set_ylabel(r'$r$', fontsize=15)

ax.scatter(s2K, s2r, label=r'$\Theta = 2$', marker='^')
ax.scatter(s3K, s3r, label=r'$\Theta = 3$', marker='^')
ax.scatter(s4K, s4r, label=r'$\Theta = 4$', marker='^')
ax.scatter(unwK, unwr, label='unweighted', marker='^')
ax.plot([0, 2.5], [1, 1], linewidth = .8, color = 'grey')

ax.legend()
plt.show()
```

```python
t = np.arange(0,10,.05)
```

```python
fig, ax = plt.subplots(1,1, figsize=(9, 6))

ax.set_ylabel(r'$f(\lambda)$')
ax.set_xlabel(r'$\lambda$')

ax.plot(t, ss.gamma.pdf(t, 2, scale=.5), label=r'$\Theta = 2$', lw=2)
ax.plot(t, ss.gamma.pdf(t, 3, scale=.3333), label=r'$\Theta = 3$', lw=2)
ax.plot(t, ss.gamma.pdf(t, 4, scale=.25), label=r'$\Theta = 4$', lw=2)

ax.set_xlim([0,3.5])
ax.legend()

plt.show()
```
